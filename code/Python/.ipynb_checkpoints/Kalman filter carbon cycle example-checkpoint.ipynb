### Carbon Cycle Model

```python
#initial states and fluxes
MCss = 3.8e6 #in 1e12 mol
MPss = 2e3
delCss = 0
Fwcarb = 34
Fv = 6
Fws = Fv
Fbo = 10
Fwo = Fbo
Fwp = 3.6e-2
Fbp = Fwp
eps = 28
d13C_volc = -5
d13C_worg = -25

#coefficient values
kws = Fws/MCss;
kbo = Fbo/MPss;
kbp = Fbp/MPss;
kwp = Fwp/MCss;
```

```python
d13C_volc + kbo*MPss/(kbo*MPss + kws*MCss + Fwcarb)*eps
```

### Simulate the model

```python
T = 10e6
# Fv_interp = interp1d(t,Fv_t)
a = -0.25
F_forcing = lambda t: Fv*(1+a*(sin(2*pi*t/T)*(1 - cos(pi*t/(0.5*T)))))
```

```python
tt = linspace(0,10e6,30)
plt.plot(tt,F_forcing(tt))
plt.title('Forcing function')
plt.xlabel('Time [Ma]')
pl
```

```python
def odeFun(y, t, F_forcing):
    Mp,Mc = y[0],y[1]
    Fv = F_forcing(t)
    dMp = kwp*Mc - kbp*Mp
    dMc = Fv + Fwo - kws*Mc - kbo*Mp
    return array([dMp,dMc])
```

```python
t = linspace(0,10e6,10000)
dt= t[1] - t[0]
M0 = [MPss,MCss]
Y = odeint(lambda y,t: odeFun(y,t,F_forcing), M0, t) 
Mp, Mc = Y[:,0],Y[:,1]
```

```python
plt.figure(figsize = (12, 6))

plt.subplot(2,2,1)
plt.plot(t,Mp,'b')
plt.ylabel(r'M$_P$')

plt.subplot(2,2,2)
plt.plot(t,Mc,'g')
plt.ylabel(r'M$_C$')

# plt.subplot(2,2,3)
# plt.plot(t,delC,'r')
# plt.ylabel(r'$\delta^{13}$C')
# plt.xlabel('t')

# plt.subplot(2,2,4)
# plt.plot(t,a,'c')
# plt.ylabel(r'a')
# plt.xlabel('t')
```
